```python
# read CSV with explicit schema definiitons
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DoubleType, DateType
```

```python
#Define the schema

schema = StructType([
    StructField(name="InvoiceNo", dataType=StringType(), nullable=True),
    StructField(name='StockCode', dataType=StringType(), nullable=True),
    StructField(name='Description', dataType=StringType(), nullable=True),
    StructField(name='Quantity', dataType=IntegerType(), nullable=True),
    StructField(name='InvoiceDate', dataType=DateType(), nullable=True),
    StructField(name='UnitPrice', dataType=DoubleType(), nullable=True),
    StructField(name='CustomerID', dataType=StringType(), nullable=True),
    StructField(name='Country', dataType=StringType(), nullable=True),
    
    
])
```

```python
##read csv
csv_file_path = "/Users/zano/Documents/course/sprk/dibimbing_spark_airflow/data/online-retail-dataset.csv"
df = spark.read.csv(csv_file_path, header=True, schema=schema)
```

```python
# Display schema of DataFrame
df.printSchema()

# Diplay schema of DataFrame
df.show(10)
```

```python
#Membaca schema secara otomatis dengan menggunakan inferschema
##read csv
csv_file_path = "/Users/zano/Documents/course/sprk/dibimbing_spark_airflow/data/online-retail-dataset.csv"
df = spark.read.csv(csv_file_path, header=True, inferschema=True)
```

```python
# Display schema of DataFrame
df.printSchema()

# Diplay schema of DataFrame
df.show(10)
```
